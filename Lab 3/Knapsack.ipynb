```python
def knapsack(weights, profit, capacity, numOfItems):
    maxProfit = [0] * (capacity + 1)
    choice = [-1] * (capacity + 1)

    for i in range(1, capacity+1):
        print(f"\nCapacity: {i}")
        for j in range(numOfItems): #loop through all items
            if(weights[j] <= i): #check if current item's weight is < current capacity
                potential_profit = maxProfit[i - weights[j]] + profit[j] #profit if we include current item
                print(f"Consider Item {j}: Weight = {weights[j]}, Profit = {profit[j]}")
                print(f"Potential Profit if this item is included: {potential_profit}\n")
                
                if potential_profit > maxProfit[i]:
                    maxProfit[i] = potential_profit
                    choice[i] = j #store index of chosen item

        if choice[i] != -1:
            chosen_item = choice[i]
            print(f"Item chosen for capacity {i}: Item {chosen_item} (Weight: {weights[chosen_item]}, Profit: {profit[chosen_item]})")

        print(f"P({i}) MaxProfit: " + " ".join(map(str, maxProfit)))

    print("\nFinal MaxProfit array:", maxProfit)
    print("Final Choice array:", choice)

    w = capacity
    items_used = []
    while(w>0 and choice[w] !=-1):
        items_used.append(choice[w])
        w -= weights[choice[w]]

    print(f"Items chosen to achieve max profit: {items_used}")
    
    return maxProfit[capacity]
                
```

```python
capacity = 14
weightsArray = [4,6,8]
profitArray = [7,6,9]
numOfItems = len(weightsArray)

maxProfit1 = knapsack(weightsArray, profitArray, capacity, numOfItems)
print("Maximum Profits for 4a: {}".format(maxProfit1))
```
